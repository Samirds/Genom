```python
idx = dataset_train[['Genetic Disorder', 'Disorder Subclass']].dropna().index
dataset_train = dataset_train.loc[idx]
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

dataset_train['Genetic Disorder'] = le.fit_transform(dataset_train['Genetic Disorder'])
dataset_train['Disorder Subclass'] = le.fit_transform(dataset_train['Disorder Subclass'])
```

```python
dataset_train.head(2)
```

# One Hot Encoding

### Data_ca 

```python
data_ca = dataset_train.drop(['Patient Id', 'Family Name', "Patient First Name", "Father's name", "Location of Institute"], axis=1)
data_ca.head(2)
```

```python
data_ca = data_ca.dropna()
print(data_ca.shape, dataset_train.shape)
```

```python
one_hot_col = [feature for feature in data_ca.columns if data_ca[feature].dtypes =='O' and feature != 'Institute Name']
one_hot_col
```

```python
data_ohe = pd.get_dummies(data_ca[one_hot_col], drop_first=True)
#data_ohe_ca = pd.concat([data_ohe, data_ca[['Genetic Disorder', 'Disorder Subclass']]], axis=1)
data_ohe.head(2)
```

# Numerical feature.dropna()

```python
dataset_train.head(2)
```

```python
data_nm = dataset_train[numerical_feature+['Genetic Disorder', 'Disorder Subclass']].dropna()
print(data_nm.shape)
data_nm.head(2)
```

# Total Data

```python
data = pd.concat([data_ohe, data_nm], axis=1)
data = data.dropna()
data = data.drop(['Test 1', 'Test 2' , 'Test 3', "Test 4", "Test 5"], axis=1)
print(data.shape)
data.head(2)
```
